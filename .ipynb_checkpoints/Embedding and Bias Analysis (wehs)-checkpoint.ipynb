### Import and Helpers

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import hiplot as hip
from sklearn.decomposition import PCA
```

```python
def distances(X1, X2):
    out = [np.sqrt(np.sum((X2-x)**2,1)) for x in X1]
    return np.array(out)

def cos_similarity(X1, X2):  # all distances between embedding vectors X1 and X2
    out = [np.sum(X2*x,1) for x in X1]
    return np.array(out)

def cumdist(v):  # x, y for cumulative density plot 
    x = np.sort(v.flatten())
    y = np.arange(len(x))/len(x)
    return x, y
```

```python
class H5Reader:
    pattern = r'C:\Daten\FaceRecognitionBias\senet50\embeddings\senet50_{net}_pytorch_{race}_normed_BGR.h5'
    race_list = ['Caucasian','Indian','Asian','African']
    net_list = ['128','256','ft']
    hdf_key = '/embeddings'
    
    def __init__(self, net):
        self.net = net
        
    def race(self, prefix):
        return list(filter(lambda s: s.lower().startswith(prefix.lower()),self.race_list))[0]
        
    def path(self, race):
        race = list(filter(lambda s: s.lower().startswith(race.lower()),self.race_list))[0]
        return self.pattern.format(net = self.net, race = race)
        
    def read(self, race_prefix, race_index = False):
        race = self.race(race_prefix)
        path = self.pattern.format(net = self.net, race = race)
        df = pd.read_hdf(path, self.hdf_key)
        if race_index:
            df['race_index'] = self.race_list.index(race)
        return df
    
    def read_all(self, N=None, select = 'head'):
        out = []
        for race in self.race_list:
            df = self.read(race, race_index=True)
            if N is None:
                out.append(df)
            elif select == 'head':
                out.append(df.head(N))
            else:
                out.append(df.tail(N))    
                
        return pd.concat(out,0).set_index('race_index').reset_index()
                
    def keys(self, race_prefix):    
        race = self.race(race_prefix)
        path = self.pattern.format(net = self.net, race = race)
        with pd.HDFStore(path) as hdf:
            out = hdf.keys()
        return out
    
r128 = H5Reader('128')
r128.keys('c')
```

## 1.1 Cosine similarities

- Cosine similarity fits the fact that embedding vectors are normalized. I.e. they lie on a hyper-sphere and therefore angles between points makes sense.
- Caucasian und African seem to be different. 
- senet50_ft does not fill the full embedding space
- Recognition rates of https://stdm.github.io/downloads/papers/ANNPR_2020.pdf are not directly visible 

```python
def cos_cdf(reader, race1, race2, N = 1000):
    X1 = reader.read(race1).values[:N]
    X2 = reader.read(race2).values[(-N):]
    d = cos_similarity(X1, X2)
    return cumdist(d)
```

```python
fig, axs = plt.subplots(5, 3, figsize=[15,20],)
#fig.suptitle('Cosine Similarities')
for inet, net in enumerate(['128','256','ft']):
    reader = H5Reader(net)
    for irace, race1 in enumerate(H5Reader.race_list):
        for race2 in H5Reader.race_list:
            x, y = cos_cdf(reader,race1, race2, N=1000)
            if race1 == race2:
                axs[irace+1, inet].plot(x, y, label=race2, lw=3)
                axs[0, inet].plot(x, y, label=race2, lw=3)
            else:
                axs[irace+1, inet].plot(x, y, label=race2, ls = '--')
        axs[irace+1,inet].set_title('Off-Diagonal, '+race1+' vs')
        axs[irace+1,inet].legend(loc='upper left')
        axs[irace+1,inet].set_xlim([-0.5,1])
    axs[0,inet].set_title('Diagonal, senet50_'+net)
    axs[0,inet].legend(loc='upper left')
    axs[0,inet].set_xlim([-0.5,1])
    
plt.savefig('cosine_similarities.png',papertype='A4',dpi=200)
plt.show()
```

### Nearest Neighbor analysis & Confusion matrix

Maybe only the nearest neighbors count -> Investigate the distribution of nearest neighbor cosine similarities. 
- NN distributions show similar hirarchy as full pairwise distributions
- Interestingly: Low average number of same-class neighbors correlates with high recognition rate (see ttps://stdm.github.io/downloads/papers/ANNPR_2020.pdf)

```python
N = 4000
rd = H5Reader('128')
df1 = rd.read_all(N=N)
df2 = rd.read_all(N=N, select='tail')
d = cos_similarity(df1.iloc[:,1:].values, df2.iloc[:,1:].values)
```

```python
resdf = df1.iloc[:,[0]]
resdf['nn_race_index'] = df2.race_index.reindex(np.argmax(d,1)).values
resdf['cos_sim'] = np.max(d,1)
resdf
```

```python
for irace, race in enumerate(H5Reader.race_list):
    x, y = cumdist(resdf.query('race_index==@irace').cos_sim.values)
    plt.plot(x, y, label = race)
plt.legend()
plt.show()
```

```python
#confusion matrix
confmat = np.zeros([4,4])
for i in range(4):
    for j in range(4):
        confmat[i,j] = np.mean(resdf.query('race_index==@i').nn_race_index == j)
confmat        
```

## 2.1. Variance analysis

```python
pca = PCA(svd_solver='full')
N=9500

hmat = np.zeros([3,4])
fig, axs = plt.subplots(1, 3, figsize=(16,5))
for inet,net in enumerate(H5Reader.net_list):
    rd = H5Reader(net)
    for irace,race in enumerate(H5Reader.race_list):
        X = rd.read(race).head(N).values
        pca.fit(X)
        var = pca.explained_variance_ratio_
        h = np.dot(var,np.arange(len(var)))
        hmat[inet,irace]=h
        axs[inet].semilogy(1 - np.cumsum(var),label=race+' '+str(np.round(h,2)))
    axs[inet].legend()
    axs[inet].set_title('senet50_'+net)
    axs[inet].set_ylim([0.01,1])
    axs[inet].set_xlim([0,150])
    axs[inet].set_xlabel('PCA component')
    if inet == 0:
        axs[inet].set_ylabel('Remaining rel. variance after component')
    
plt.show()
```

### Variance spread and recognition rate

PCA calculates normalized variance $p_i$ such that $\sum p_i = 1$. Here I define a variance spread as

$$ h = \sum_i i * p_i $$

- Bigger h means that variance spreads over more components, i.e. more dimensions. 
- Below: h is calculated with a seperate pca for each class (this could also be changed).
- Fit is done with a fit-paremeter per model

```python
# recognition rate
recrate=[[0.8906,0.8531,0.8310,0.7998],
        [0.8787,0.8265,0.7981,0.7597],
        [0.8799,0.8472,0.8305,0.7959]]

recrate_fit = hmat*np.mean(recrate/hmat,1,keepdims=1)

for inet, net in enumerate(H5Reader.net_list):
    plt.scatter(hmat[inet],recrate[inet],label='paper, ' + net)
    plt.plot(hmat[inet],recrate_fit[inet],label='fit, '+net)
plt.legend()
plt.xlabel('variance spread')
plt.ylabel('recognition rate')
plt.title('Predicting Recognition Rates')
plt.show()
```

### Distribution functions of PCA prejections 

Here PCA was done on all classes

```python
N = 9500
pca_dict = {}
for net in H5Reader.net_list:
    df = H5Reader(net).read_all()
    X = df.values[:,1:]
    pca = PCA(n_components=10)
    Xp = pca.fit_transform(X)
    pca_dict[net] = pd.concat([df.drop(columns=df.columns[1:]),pd.DataFrame(data=Xp)],1) 
```

```python
Npca = 8
fig, axs = plt.subplots(Npca, 3, figsize=[15,20],)
#fig.suptitle('Cosine Similarities')
for inet, net in enumerate(H5Reader.net_list):
    for ipca in range(Npca):
        for irace, race in enumerate(H5Reader.race_list):
            x, y = cumdist(pca_dict[net].query('race_index==@irace')[ipca].values)
            axs[ipca, inet].plot(x,y,label=race)
        if ipca == 0:
            axs[ipca, inet].set_title('senet50_'+net)
        axs[ipca, inet].legend(loc='upper left')
        axs[ipca,inet].set_xlim([-0.6,0.6])
        if inet == 0:
            axs[ipca,inet].set_ylabel('PCA component '+str(ipca + 1))
    
plt.savefig('pca.png',papertype='A4',dpi=200)
plt.show()
```
